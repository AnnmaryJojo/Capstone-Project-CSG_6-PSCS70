```python
# Duration in seconds
duration = librosa.get_duration(y=y, sr=sr)
print(f"Duration: {duration:.2f} sec")

# Root Mean Square (RMS) energy
rms = librosa.feature.rms(y=y)
print(f"Average RMS energy: {np.mean(rms):.4f}")

# Zero Crossing Rate (ZCR) - indicates voiced/unvoiced segments
zcr = librosa.feature.zero_crossing_rate(y)
print(f"Average ZCR: {np.mean(zcr):.4f}")
```

```python
# Install required libraries
!pip install moviepy pydub webrtcvad numpy

from moviepy.editor import VideoFileClip, AudioFileClip
from pydub import AudioSegment
import webrtcvad
import numpy as np
from google.colab import files

# Upload original video and translated audio
uploaded = files.upload()  # upload original_video.mp4 + translated_audio.mp3
video_file = [f for f in uploaded.keys() if f.endswith(".mp4")][0]
translated_audio_file = [f for f in uploaded.keys() if f.endswith(".mp3")][0]

# Load audio from original video and process with pydub
video = VideoFileClip(video_file)
original_audio = video.audio
original_audio_file = "temp_original.wav"
original_audio.write_audiofile(original_audio_file) # Write as default for pydub loading

# Load audio segments and convert for VAD
orig_audio = AudioSegment.from_wav(original_audio_file)
# Convert to 16-bit mono PCM at 16000 Hz for webrtcvad
orig_audio = orig_audio.set_frame_rate(16000).set_channels(1).set_sample_width(2)

trans_audio = AudioSegment.from_mp3(translated_audio_file)

# Setup VAD
vad = webrtcvad.Vad(1)  # 1=low aggressiveness
frame_ms = 30
frame_bytes = int(orig_audio.frame_rate * frame_ms / 1000) * orig_audio.sample_width # Calculate frame_bytes based on sample width

# Convert AudioSegment to bytes for VAD
samples = np.array(orig_audio.get_array_of_samples())
speech_frames = []

# Process frames for VAD
# Ensure frame length is a multiple of sample width
frame_length = int(orig_audio.frame_rate * frame_ms / 1000) * orig_audio.sample_width
for i in range(0, len(samples) * orig_audio.sample_width // orig_audio.sample_width, frame_length // orig_audio.sample_width):
    frame = samples[i:i + frame_length // orig_audio.sample_width]
    if len(frame) * orig_audio.sample_width < frame_length:
        break
    # Convert frame to bytes
    frame_bytes_data = frame.tobytes()
    is_speech = vad.is_speech(frame_bytes_data, sample_rate=orig_audio.frame_rate)
    speech_frames.append(is_speech)

# Simple logic: detect first speech start
if True in speech_frames:
    first_speech_index = speech_frames.index(True)
    speech_start_ms = first_speech_index * frame_ms
else:
    speech_start_ms = 0

print(f"Detected speech start at {speech_start_ms} ms")

# Insert translated audio at detected speech start
combined_audio = AudioSegment.silent(duration=speech_start_ms) + trans_audio

# Export final audio
aligned_audio_file = "aligned_translated_audio.wav"
combined_audio.export(aligned_audio_file, format="wav")

# Merge with original video
new_audio_clip = AudioFileClip(aligned_audio_file)
# Ensure the new audio duration doesn't exceed the video duration
if new_audio_clip.duration > video.duration:
    new_audio_clip = new_audio_clip.subclip(0, video.duration)

final_video = video.set_audio(new_audio_clip)
output_file = "video_with_aligned_audio.mp4"
final_video.write_videofile(output_file, codec="libx264", audio_codec="aac")

# Download final video
files.download(output_file)
```

```python
!pip install moviepy jiwer openai-whisper

from moviepy.editor import VideoFileClip
from jiwer import wer, cer
import whisper
from google.colab import files
import os

# Upload original and final video
print("Upload the original video file (e.g., your_original_video.mp4) and the final lip-synced video file (e.g., video_lipsynced_Kannada.mp4):")
uploaded = files.upload()

# Identify files - look for .mp4 files among uploaded
mp4_files = [f for f in uploaded.keys() if f.endswith(".mp4")]

if len(mp4_files) < 2:
    print("❌ Error: Please upload both the original video file and the final lip-synced video file (.mp4 format).")
else:
    # Assuming the user uploads original first, then the final one.
    # This is an assumption and might need adjustment based on user's upload behavior.
    original_video_file = mp4_files[0]
    final_video_file = mp4_files[1]

    print(f"Assuming original video: {original_video_file}")
    print(f"Assuming final video: {final_video_file}")

    try:
        # Load video clips
        original_video = VideoFileClip(original_video_file)
        final_video = VideoFileClip(final_video_file)

        # Extract audio as temporary WAV
        original_audio = original_video.audio
        final_audio = final_video.audio

        # Trim audio to video duration if necessary
        if original_audio.duration > original_video.duration:
            original_audio = original_audio.subclip(0, original_video.duration)
            print("Trimmed original audio to match video duration.")

        if final_audio.duration > final_video.duration:
            final_audio = final_audio.subclip(0, final_video.duration)
            print("Trimmed final audio to match video duration.")

        original_audio_file = "original_audio_for_wer.wav" # Use a distinct name
        original_audio.write_audiofile(original_audio_file, fps=16000, codec="pcm_s16le")

        final_audio_file = "final_audio_for_wer.wav" # Use a distinct name
        final_audio.write_audiofile(final_audio_file, fps=16000, codec="pcm_s16le")

        # Transcribe audios using Whisper
        # Check if model is already loaded, otherwise load it
        if 'model' not in locals():
          model = whisper.load_model("small")
          print("Whisper model loaded.")
        else:
          print("Using already loaded Whisper model.")


        ref_text = model.transcribe(original_audio_file)["text"]
        pred_text = model.transcribe(final_audio_file)["text"]

        # Compute transcription-based metrics
        word_error = wer(ref_text, pred_text)
        char_error = cer(ref_text, pred_text)

        print(f"Word Error Rate (WER) between original & final audio transcription: {word_error:.2f}")
        print(f"Character Error Rate (CER) between original & final audio transcription: {char_error:.2f}")

    except Exception as e:
        print(f"An error occurred during processing: {e}")

    finally:
        # Clean up temporary audio files
        if os.path.exists(original_audio_file):
            os.remove(original_audio_file)
        if os.path.exists(final_audio_file):
            os.remove(final_audio_file)
        print("Cleaned up temporary audio files.")
```

```python
# Install required libraries
!pip install nltk jiwer whisper

import whisper
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.meteor_score import meteor_score
from jiwer import wer, cer
from moviepy.editor import VideoFileClip
from google.colab import files

# Step 1: Upload final dubbed videos and reference translations
print("Upload all final dubbed videos (mp4) and reference text files (txt)")
uploaded = files.upload()

# Detect MP4 and TXT files
videos = {f.split("_")[-1].split(".")[0]: f for f in uploaded.keys() if f.endswith(".mp4")}
references = {f.split("_")[-1].split(".")[0]: f for f in uploaded.keys() if f.endswith(".txt")}

# Step 2: Load Whisper model
model = whisper.load_model("small")

# Step 3: Loop through each language
for lang in videos.keys():
    print(f"\n=== Evaluating {lang} ===")

    video_file = videos[lang]
    reference_file = references.get(lang)
    if reference_file is None:
        print(f"❌ No reference text found for {lang}, skipping...")
        continue

    # Extract audio from final video
    video_clip = VideoFileClip(video_file)
    audio_file = f"temp_audio_{lang}.wav"
    video_clip.audio.write_audiofile(audio_file, fps=16000, codec="pcm_s16le")

    # Transcribe dubbed audio
    result = model.transcribe(audio_file)
    transcribed_text = result["text"]

    # Load reference translation
    with open(reference_file, "r", encoding="utf-8") as f:
        reference_text = f.read()

    # Compute WER / CER
    word_error = wer(reference_text, transcribed_text)
    char_error = cer(reference_text, transcribed_text)

    # Compute BLEU and METEOR
    reference_words = reference_text.split()
    predicted_words = transcribed_text.split()
    bleu = sentence_bleu([reference_words], predicted_words)
    meteor = meteor_score([reference_text], transcribed_text)

    # Print results
    print(f"Word Error Rate (WER): {word_error:.2f}")
    print(f"Character Error Rate (CER): {char_error:.2f}")
    print(f"BLEU Score: {bleu:.2f}")
    print(f"METEOR Score: {meteor:.2f}")
```

```python
!pip install jiwer nltk moviepy openai-whisper

import whisper
from moviepy.editor import VideoFileClip
from jiwer import wer, cer
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.meteor_score import meteor_score

# File paths (use your uploaded names)
english_video = "Do not Trust Anybody  - Panchatantra In English - Moral Stories for Kids - Children's Fairy Tales.mp4"
kannada_video = "merged_video.mp4"

# Step 1: Extract audio
english_audio = "english_audio.wav"
kannada_audio = "kannada_audio.wav"

VideoFileClip(english_video).audio.write_audiofile(english_audio, fps=16000, codec="pcm_s16le")
VideoFileClip(kannada_video).audio.write_audiofile(kannada_audio, fps=16000, codec="pcm_s16le")

# Step 2: Load Whisper model
model = whisper.load_model("small")

# Transcribe both
english_trans = model.transcribe(english_audio)["text"]
kannada_trans = model.transcribe(kannada_audio)["text"]

print("\n--- Transcriptions ---")
print("English:", english_trans[:200], "...")
print("Kannada:", kannada_trans[:200], "...")

# Step 3: Compute metrics
# WER & CER (treating English transcription as reference, Kannada as prediction)
word_error = wer(english_trans, kannada_trans)
char_error = cer(english_trans, kannada_trans)

# BLEU & METEOR (works better if both are in same language, but we check overlap)
reference = english_trans.split()
prediction = kannada_trans.split()
bleu = sentence_bleu([reference], prediction)
meteor = meteor_score([english_trans], kannada_trans)

print("\n--- Evaluation Results ---")
print(f"Word Error Rate (WER): {word_error:.2f}")
print(f"Character Error Rate (CER): {char_error:.2f}")
print(f"BLEU Score: {bleu:.2f}")
print(f"METEOR Score: {meteor:.2f}")
```

```python
# ===========================
# Step 1: Install dependencies
# ===========================
!pip install -q git+https://github.com/openai/whisper.git
!pip install jiwer nltk moviepy
from google.colab import files

import whisper
from moviepy.editor import VideoFileClip
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# ===========================
# Step 2: Define video paths
# ===========================
# Upload these before running: original English video + Kannada dubbed video
print("Please upload the original English video and the dubbed video:")
uploaded = files.upload()

# Find the uploaded video files
uploaded_files = list(uploaded.keys())
original_video_path = None
dubbed_video_path = None

for file in uploaded_files:
  if "English" in file and file.endswith(".mp4"):
    original_video_path = file
  elif "merged_video" in file and file.endswith(".mp4"):
    dubbed_video_path = file

if not original_video_path or not dubbed_video_path:
    print("❌ Error: Could not find both the original English video and the merged video among uploaded files.")
else:
    print(f"Original English video: {original_video_path}")
    print(f"Dubbed video: {dubbed_video_path}")

    # ===========================
    # Step 3: Extract audio
    # ===========================
    orig_audio_path = "original_audio.wav"
    dub_audio_path  = "dubbed_audio.wav"

    try:
        VideoFileClip(original_video_path).audio.write_audiofile(orig_audio_path, codec="pcm_s16le")
        VideoFileClip(dubbed_video_path).audio.write_audiofile(dub_audio_path, codec="pcm_s16le")

        # ===========================
        # Step 4: Load Whisper + Transcribe
        # ===========================
        model = whisper.load_model("small")

        orig_result = model.transcribe(orig_audio_path, language="en")
        dub_result  = model.transcribe(dub_audio_path)  # auto-detects Kannada

        orig_text = orig_result["text"].strip()
        dub_text  = dub_result["text"].strip()

        print("\nOriginal English Transcript:\n", orig_text[:300], "...")
        print("\nKannada Dub Transcript:\n", dub_text[:300], "...")

        # ===========================
        # Step 5: Compute BLEU score
        # ===========================
        smoothing = SmoothingFunction().method1
        reference = [orig_text.split()]
        candidate = dub_text.split()

        bleu_score = sentence_bleu(reference, candidate, smoothing_function=smoothing)

        print("\n=== Final BLEU Score ===")
        print("BLEU:", round(bleu_score * 100, 2), "%")

    except Exception as e:
        print(f"An error occurred during processing: {e}")
```
